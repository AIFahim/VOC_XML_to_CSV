```python
import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET


import torch
from IPython.display import Image, clear_output  # to display images
#from utils.google_utils import gdrive_download  # to download models/datasets


def xml_to_csv(path):
    xml_list = []
    for xml_file in glob.glob(path):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            bbx = member.find('bndbox')
            xmin = int(bbx.find('xmin').text)
            ymin = int(bbx.find('ymin').text)
            xmax = int(bbx.find('xmax').text)
            ymax = int(bbx.find('ymax').text)
            #label = member.find('name').text

            value = (root.find('filename').text,
                     int(root.find('size')[0].text),
                     int(root.find('size')[1].text),
                     #label,
                     [xmin,
                     ymin,
                     xmax,
                     ymax],
                     root.find('source')[0].text
                     )
            xml_list.append(value)
    column_name = ['image_id', 'width', 'height',
                   'bbox' , 'source']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df
```

```python
datasets = 'Dataset_all_jpg'
path = os.path.join(os.getcwd(), datasets , "*.xml")
all_xml = glob.glob(path)
len(all_xml)
```

```python
# 231.xml has error in parsing as it was not stored in VOC XML format
df = pd.DataFrame()
for i in all_xml:
    try:
        xml_df  = xml_to_csv(i)
        xml_df["image_id"] = i
        df = df.append(xml_df, ignore_index = True)
    except:
        continue
```

```python
# EOL error if you put \ at the end of a string. Replacing it with \\ will work fine
left = "/content/Dataset_all_jpg/" 
df['image_id'] = df['image_id'].map(lambda x: x.lstrip(left).rstrip('.xml')) + ".jpg"
```

```python
df.drop("image_id",axis=1, inplace = True)
```
